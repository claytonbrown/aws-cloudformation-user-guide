```python
# LOAD DEPENDENCIES 
import logging
import sys
from colorlog import ColoredFormatter
from cog.torque import Graph
g = Graph("cfn")
```

```python
# SETUP LOGGING

log = logging.getLogger()
verbosity = logging.INFO
logging.basicConfig(filename='debug.log', level=verbosity)
formatter_console_colour = "%(log_color)s%(levelname)-8s %(module)s:%(lineno)d- %(name)s%(reset)s %(blue)s%(message)s"
console_handler = logging.StreamHandler(sys.stdout)
console_handler.setLevel(verbosity)
colored_formatter = ColoredFormatter(
    formatter_console_colour,
    datefmt='%H:%M:%S ',
    reset=True,
    log_colors={
        'DEBUG': 'white',
        'INFO': 'green',
        'WARNING': 'yellow',
        'ERROR': 'orange',
        'CRITICAL': 'red',
    }
)
console_handler.setFormatter(colored_formatter)
log.addHandler(console_handler)
log.debug("logging initialized")
```

```python
# IMPORT DATA SOURCES 
import json
import glob 

resource_rules = {}
aws_properties = {}
cfn_global = {}

# LOAD CFN SCHEMAs
cfn_schema = json.load(open('CloudFormationResourceSpecification.json','r'))
cfn_sources = glob.glob("aws-cfn-resource-specs/specs/*/CloudFormationResourceSpecification.json")

# INGEST ALL SCHEMAS
for file in cfn_sources:
    # print(file)
    region = file.split('/CloudFormationResourceSpecification')[0].split('specs/')[-1]
    log.info("Loading cfn spec for Region: %s" % (region))
    with open(file, 'r') as f:
        try:
            cfn_global[region] = json.loads(f.read())
            log.info("Processed Region: %s " % (file))
        except Excpetion as e:
            log.warning(e)

log.info("Schemas: %s" % (
    json.dumps(
        sorted(
            list(
                cfn_global.keys()
            )
        )
    )
    )
)






```

```python
# INGEST DOCUMENTATION FOR ALLOWED VALUES etc
aws_doc_sources = glob.glob('doc_source/aws-properties-*.md.properties.json')
count = 0
for doc in aws_doc_sources:
    # print(doc)
    log.info('Reading: %s' % (doc))
    with open(doc, 'r') as f:
        property_key = doc.split(".md.properties.json")[0].split("doc_source/aws-properties-")[-1].replace('-','.')        
        try:
            data = json.loads(f.read())
            aws_properties[property_key] = { "Properties": sorted(list(data.keys())) }
            for k, v in data.items():
                aws_properties[k.lower()] = v
                count +=1 
        except Exception as e:
            log.error(e)
            
log.info("Ingested %s documentation properties]\n%s" % 
         (
            count, 
            sorted(
                list(
                    aws_properties.keys()                
                )
            )
        )
)


```

```python
# PROCESS SCHEMA elements 

schema_key = {}
cfn_selectors = {}

g.put("cfn","schema","Required")
g.put("cfn","schema","NotRequired")
g.put("cfn","schema","Mutable")
g.put("cfn","schema","Immutable")

# Schemas --> Regions 
for region in cfn_global.keys():
    g.put('cfn',"Region",region)

# For each ResourceType --> Regions
for resource_name, resource_properties in cfn_schema["ResourceTypes"].items():
    key = resource_name #.replace('::','.')
    schema_key[key] = resource_name
    log.info("%s --> %s" % (key, resource_name))

    g.put("cfn","Resource",resource_name)
    
    # Resource Region Availability
    for region in cfn_global.keys():
        g.put('cfn',"Region",region)
        if resource_name in cfn_global[region]["ResourceTypes"]:
            g.put(resource_name,"Resource",region)
            if "Regions" not in cfn_schema["ResourceTypes"][resource_name]:
                cfn_schema["ResourceTypes"][resource_name]["Regions"] = [region]                
            else:
                # only add once as in inner loop and not a set
                if region not in regioncfn_schema["ResourceTypes"][resource_name]["Regions"]:
                    cfn_schema["ResourceTypes"][resource_name]["Regions"].append(region)
    
    # Resource Properties 
    if "Properties" in cfn_schema["ResourceTypes"][resource_name]:
        these_properties = cfn_schema["ResourceTypes"][resource_name]["Properties"]
        for resource_property_name, resource_property in these_properties.items():
            # resource_property 
            
            if "Documentation"in resource_property:
                doc_source = resource_property["Documentation"].split('#cfn-')[-1].replace('-','.')
                # doc_key = "%s.%s" % (doc_source, )                
                print("DocSourceCheck\t\t%s : %s " % (doc_source, doc_source in aws_properties))
                print("CfnPropertiesCheck\t%s : %s " % (resource_property_name, resource_property_name in cfn_schema["PropertyTypes"]))
                if doc_source in aws_properties:
                    resource_property["Shape"] = {} #aws_properties[doc_source]["Properties"]
                    for type_property in aws_properties[doc_source]["Properties"]:
                        print("Type Property: %s" % (type_property))
                        type_property_key = type_property #"%s.%s" % ()
                        resource_property["Shape"][type_property_key] = aws_properties[type_property_key.lower()]
                        print("Added: %s" % (type_property_key))
                        
                
            if "PrimitiveType" not in resource_property and 'Shape' not in resource_property:
                
                if resource_property["Type"] in cfn_schema["PropertyTypes"]:
                    type_data = cfn_schema["PropertyTypes"][resource_property["Type"]]
                    print("Found Type in CFN Schema: %s" % (json.dumps(type_data, indent=4)))
                    
                elif resource_property["Type"] == "List" and "ItemType" in resource_property and resource_property["ItemType"] in cfn_schema["PropertyTypes"]:
                    type_data = cfn_schema["PropertyTypes"][resource_property["ItemType"]]
                    print("Found ItemType in CFN Schema: %s" % (json.dumps(type_data, indent=4)))
                else:
                    print("TODO: '%s': %s" % (resource_property, json.dumps(resource_property, indent=4)))
                

    # create CNF guard 2 resource selector
    cfn_selectors[resource_name] = {}

    if "Properties" in resource_properties:
        
        for resource_property_name, resource_property_detail in resource_properties["Properties"].items():
            property_key = "%s.%s" % (key, resource_property_name)
            property_ref = "%s.%s" % (resource_name, resource_property_name)
            schema_key[property_key] = property_ref
            log.info("%s --> %s" % (property_key, property_ref))

            cfn_selectors[resource_name][resource_property_name]= {}

            g.put(resource_name,"has",resource_property_name)
    print("Processed: %s" % (resource_name))
```

```python
# INGEST DOCUMENTATION FOR ALLOWED VALUES etc
aws_doc_sources = glob.glob('doc_source/aws-properties-*.md.properties.json')
count = 0
for doc in aws_doc_sources:
    # print(doc)
    log.info('Reading: %s' % (doc))
    with open(doc, 'r') as f:
        try:
            data = json.loads(f.read())
            for k, v in data.items():
                aws_properties[k.lower()] = v
                count +=1 
        except Exception as e:
            log.error(e)
            
print("Ingested %s documentation properties]\n%s" % 
         (
            count, 
            sorted(
                list(
                    aws_properties.keys()                
                )
            )
        )
)

```
